```python
plt.figure(figsize=(5,5))
sns.countplot(x='Pregnancies',hue='Outcome',data=df)
plt.title("Count of Diabetic Patients")
plt.show()
```

```python
plt.figure(figsize=(5,5))
df.plot(kind='scatter',x='Pregnancies',y='Insulin')
plt.title("Count of Insulin Level")
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.countplot(x='Age',hue='Outcome',data=df)
plt.title("Count of Diabetic Patient by Age")
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.barplot(y='SkinThickness',hue='Outcome',data=df)
plt.title("SkinThickness Analysis")
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x='Age',y='Glucose',hue='Outcome',data=df)
plt.title("Glucose_Level by Age")
plt.show()
```
